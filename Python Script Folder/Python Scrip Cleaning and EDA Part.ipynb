# Exploratrey Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q.1 Find the top category by video count and top tranding country by video count.

```python
plt.figure(figsize=(10,8))
sns.countplot(y="video_category_id",data=df,order=df["video_category_id"].value_counts().index)
plt.title("Top Category by Video Count")
plt.show()
```

```python
plt.figure(figsize=(10,20))
sns.countplot(y="video_trending_country",data=df,order=df["video_trending_country"].value_counts().index)
plt.title("Top Tranding Country by Video Count")
plt.show()
```

```python
# Q.2 Find the channel Country by video count
plt.figure(figsize=(20,5))
sns.countplot(y="channel_country",data=df,order=df["channel_country"].value_counts().index)
plt.title("Top Chananl Country by Video Count")
plt.show()
```

# Analyzing Relationship

## Q.3 Find the relationship Between video_view_count, video_like_count and video_comment_count

```python
plt.figure(figsize=(25,5))
plt.subplot(1,3,1)
sns.scatterplot(data=df,x="video_view_count",y="video_like_count")
plt.xlabel("Viwe_count")
plt.ylabel("Like_count")
plt.title("relationship Between video_view_count & video_like_count in(Log Scale)")

plt.subplot(1,3,2)
sns.scatterplot(data=df,x="video_view_count",y="video_comment_count")
plt.xlabel("Viwe_count")
plt.ylabel("Comment_count")
plt.title("relationship Between video_view_count & video_comment_count in (Log Scale)")

plt.subplot(1,3,3)
sns.scatterplot(data=df,x="channel_video_count",y="channel_subscriber_count")
plt.xlabel("Channel_video_count")
plt.ylabel("Channel_subscriber_count")
plt.title("relationship Between Channel_video_count & Channel_subscriber_count in (Log Scale)")

plt.tight_layout()
plt.show()
```

## Q.4 Find the correlation between numerical data

```python
Corelation_videos_matrix=df[["video_view_count","video_like_count","video_comment_count"]].corr()
Corelation_channel_matrix=df[["channel_view_count","channel_subscriber_count","channel_video_count"]].corr()

plt.figure(figsize=(22,5))
plt.subplot(1,2,1)
sns.heatmap(Corelation_videos_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation matrix of video viwe, like and comment")

plt.subplot(1,2,2)
sns.heatmap(Corelation_channel_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation matrix of channel viwe, subscriber and video")

plt.show()
```
